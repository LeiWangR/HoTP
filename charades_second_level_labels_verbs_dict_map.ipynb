```python
import torch
import torch.nn as nn

m = nn.Sigmoid()
loss = nn.BCELoss()
input = torch.randn(3, requires_grad=True)
target = torch.empty(3).random_(2)
```

```python
print(input)
```

```python
print(target)
```

```python
output = loss(m(input), target)
output.backward()
```

```python
import pandas as pd

mapping_txt = pd.read_csv('/Users/wanglei/Downloads/Charades/Charades_v1_mapping.txt', delim_whitespace=True, header = None)
```

```python
c_label = []
v_label = []

for i in range(mapping_txt.shape[0]):
    # print(i)
    class_l = int(mapping_txt.iloc[i, 0][1:4])
    class_v = int(mapping_txt.iloc[i, 2][1:4])
    c_label.append(class_l)
    v_label.append(class_v)
```

```python
new_dict = {k: v for k, v in zip(c_label, v_label)}
```
