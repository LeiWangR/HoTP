```python
import csv
import pandas as pd
import numpy as np
import math

with open('Charades_v1_train.csv') as f:
    reader = csv.DictReader(f)
    sum_x = 0
    all_actions = []
    for row in reader:
        video_name = row['id']
        actions = row['actions'].split(';')
        length_v = row['length']
        if not actions == ['']:
            # print(video_name, '----------', actions, '***')
            actions_length = len(actions)
            sum_x = sum_x + actions_length
            for t in range(actions_length):
                each_action = video_name + ' ' + actions[t] + ' ' + length_v
                all_actions.append(each_action)
                each_action = each_action.split()
                # print(each_action)
# print(all_actions)


# big_list = []
new_big_list = []
for ii in range(len(all_actions)):
    each_action = all_actions[ii].split()
    # 
    action_label = each_action[1]
    label = int(action_label[1:4])
    if float(each_action[2]) < float(each_action[3]):
        new_each_action = [each_action[0], label, math.ceil(float(each_action[2]) * 24), math.ceil(float(each_action[3]) * 24), math.ceil(float(each_action[3]) * 24) - math.ceil(float(each_action[2]) * 24) + 1, float(each_action[2]) / float(each_action[4]), float(each_action[3]) / float(each_action[4])]
    else:
        new_each_action = [each_action[0], label, math.ceil(float(each_action[3]) * 24), math.ceil(float(each_action[2]) * 24), math.ceil(float(each_action[2]) * 24) - math.ceil(float(each_action[3]) * 24) + 1, float(each_action[3]) / float(each_action[4]), float(each_action[2]) / float(each_action[4])]
    # big_list.append(each_action)
    
    new_big_list.append(new_each_action)
# print(new_big_list)
myarray = np.asarray(new_big_list)
df_myarray = pd.DataFrame(myarray)
df_myarray.head()
df_myarray.to_csv('charades_info_list_tr.csv', index = False, header = False)
```

# Training and Testing Video Frame Counts

## ------ Training

```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np

df_myarray = pd.read_csv('charades_info_list_tr.csv', header = None)
frames_vector = df_myarray.iloc[:, 4].values
results = list(map(int, frames_vector))

print('Mean: ', math.ceil(np.mean(results)), ' Min: ', np.min(results), ' Max: ', np.max(results))

plt.hist(results, bins= np.arange(20, 3640, 50))
plt.xlabel('Frames')
plt.ylabel('Counts')
plt.title('Frame Counts for Training Set')
plt.show()
```

## ------ Testing

```python
df_myarray = pd.read_csv('charades_info_list_te.csv', header = None)
frames_vector = df_myarray.iloc[:, 4].values
results = list(map(int, frames_vector))

print('Mean: ', math.ceil(np.mean(results)), ' Min: ', np.min(results), ' Max: ', np.max(results))

plt.hist(results, bins= np.arange(20, 3640, 50))
plt.xlabel('Frames')
plt.ylabel('Counts')
plt.title('Frame Counts for Test Set')
plt.show()
```
