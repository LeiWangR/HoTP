# Tr-Te split functions

# for csv files

```python
import pandas as pd

def tr_te_split(df_data):
    
    mpii_info = pd.read_csv(df_data, header = None)
    print('--- name: ', df_data, '|', 'shape: ', mpii_info.shape)
    
    mpii_p1 = mpii_info[mpii_info.iloc[:, 0] == 11]
    mpii_p2 = mpii_info[mpii_info.iloc[:, 0] == 12]
    mpii_p3 = mpii_info[mpii_info.iloc[:, 0] == 13]
    mpii_p4 = mpii_info[mpii_info.iloc[:, 0] == 14]
    mpii_p5 = mpii_info[mpii_info.iloc[:, 0] == 15]
    
    mpii_q1 = mpii_info[mpii_info.iloc[:, 0] == 8]
    mpii_q2 = mpii_info[mpii_info.iloc[:, 0] == 10]
    mpii_q3 = mpii_info[mpii_info.iloc[:, 0] == 16]
    mpii_q4 = mpii_info[mpii_info.iloc[:, 0] == 17]
    mpii_q5 = mpii_info[mpii_info.iloc[:, 0] == 18]
    mpii_q6 = mpii_info[mpii_info.iloc[:, 0] == 19]
    mpii_q7 = mpii_info[mpii_info.iloc[:, 0] == 20]
    
    mpii_tr_sp1 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q2, mpii_q3, mpii_q4, mpii_q5, mpii_q6, mpii_q7], axis = 0)
    
    mpii_tr_sp2 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q1, mpii_q3, mpii_q4, mpii_q5, mpii_q6, mpii_q7], axis = 0)
    
    mpii_tr_sp3 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q1, mpii_q2, mpii_q4, mpii_q5, mpii_q6, mpii_q7], axis = 0)
    
    mpii_tr_sp4 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q1, mpii_q2, mpii_q3, mpii_q5, mpii_q6, mpii_q7], axis = 0)
    
    mpii_tr_sp5 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q1, mpii_q2, mpii_q3, mpii_q4, mpii_q6, mpii_q7], axis = 0)
    
    mpii_tr_sp6 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q1, mpii_q2, mpii_q3, mpii_q4, mpii_q5, mpii_q7], axis = 0)
    
    mpii_tr_sp7 = pd.concat([mpii_p1, mpii_p2, mpii_p3, mpii_p4, mpii_p5, 
                         mpii_q1, mpii_q2, mpii_q3, mpii_q4, mpii_q5, mpii_q6], axis = 0)
    
    
    mpii_te_sp1 = mpii_q1
    mpii_te_sp2 = mpii_q2
    mpii_te_sp3 = mpii_q3
    mpii_te_sp4 = mpii_q4
    mpii_te_sp5 = mpii_q5
    mpii_te_sp6 = mpii_q6
    mpii_te_sp7 = mpii_q7
    
    mpii_tr_sp1.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp1.csv', header = None, index = None)
    mpii_tr_sp2.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp2.csv', header = None, index = None)
    mpii_tr_sp3.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp3.csv', header = None, index = None)
    mpii_tr_sp4.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp4.csv', header = None, index = None)
    mpii_tr_sp5.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp5.csv', header = None, index = None)
    mpii_tr_sp6.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp6.csv', header = None, index = None)
    mpii_tr_sp7.to_csv('tr_te_split/' + df_data[:-4] + '_tr_sp7.csv', header = None, index = None)
    
    mpii_te_sp1.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp1.csv', header = None, index = None)
    mpii_te_sp2.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp2.csv', header = None, index = None)
    mpii_te_sp3.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp3.csv', header = None, index = None)
    mpii_te_sp4.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp4.csv', header = None, index = None)
    mpii_te_sp5.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp5.csv', header = None, index = None)
    mpii_te_sp6.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp6.csv', header = None, index = None)
    mpii_te_sp7.to_csv('tr_te_split/' + df_data[:-4] + '_te_sp7.csv', header = None, index = None)
    
    print('*sp1--tr: ', mpii_tr_sp1.shape, '|', '--te: ', mpii_te_sp1.shape)
    print('*sp2--tr: ', mpii_tr_sp2.shape, '|', '--te: ', mpii_te_sp2.shape)
    print('*sp3--tr: ', mpii_tr_sp3.shape, '|', '--te: ', mpii_te_sp3.shape)
    print('*sp4--tr: ', mpii_tr_sp4.shape, '|', '--te: ', mpii_te_sp4.shape)
    print('*sp5--tr: ', mpii_tr_sp5.shape, '|', '--te: ', mpii_te_sp5.shape)
    print('*sp6--tr: ', mpii_tr_sp6.shape, '|', '--te: ', mpii_te_sp6.shape)
    print('*sp7--tr: ', mpii_tr_sp7.shape, '|', '--te: ', mpii_te_sp7.shape)
```

```python
import glob
csv_list = sorted(glob.glob('*.csv'))

for ii in range(len(csv_list)):
    csv_file = csv_list[ii]
    tr_te_split(csv_file)
```

# For features (to be uploaded to cluster)

```python
import torch
import numpy as np

raw_dir = 'Raw_400D/'
def tr_te_feature_split(tensor_file, df_data):
    save_dir = 'tr_te_raw_400D/'
    tensor_file = raw_dir + tensor_file
    # bow_tensor = torch.load('mpii_bow_sketched_4000D.pth', map_location={'cuda:0': 'cpu'})
    feat_tensor = torch.load(tensor_file)
    print('--- name: ', tensor_file, '|', 'shape: ', feat_tensor.shape)
    mpii_info = pd.read_csv(df_data, header = None)
    print('--- name: ', df_data, '|', 'shape: ', mpii_info.shape)
    
    a1 = (mpii_info.iloc[:, 0] == 11).values
    i1 = np.where(a1)[0]
    a2 = (mpii_info.iloc[:, 0] == 12).values
    i2 = np.where(a2)[0]
    a3 = (mpii_info.iloc[:, 0] == 13).values
    i3 = np.where(a3)[0]
    a4 = (mpii_info.iloc[:, 0] == 14).values
    i4 = np.where(a4)[0]
    a5 = (mpii_info.iloc[:, 0] == 15).values
    i5 = np.where(a5)[0]

    ba1 = (mpii_info.iloc[:, 0] == 8).values
    bi1 = np.where(ba1)[0]
    ba2 = (mpii_info.iloc[:, 0] == 10).values
    bi2 = np.where(ba2)[0]
    ba3 = (mpii_info.iloc[:, 0] == 16).values
    bi3 = np.where(ba3)[0]
    ba4 = (mpii_info.iloc[:, 0] == 17).values
    bi4 = np.where(ba4)[0]
    ba5 = (mpii_info.iloc[:, 0] == 18).values
    bi5 = np.where(ba5)[0]
    ba6 = (mpii_info.iloc[:, 0] == 19).values
    bi6 = np.where(ba6)[0]
    ba7 = (mpii_info.iloc[:, 0] == 20).values
    bi7 = np.where(ba7)[0]
    
    ind_tr_s1 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi2, bi3, bi4, bi5, bi6, bi7), axis=0)
    ind_tr_s2 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi1, bi3, bi4, bi5, bi6, bi7), axis=0)
    ind_tr_s3 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi1, bi2, bi4, bi5, bi6, bi7), axis=0)
    ind_tr_s4 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi1, bi2, bi3, bi5, bi6, bi7), axis=0)
    ind_tr_s5 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi1, bi2, bi3, bi4, bi6, bi7), axis=0)
    ind_tr_s6 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi1, bi2, bi3, bi4, bi5, bi7), axis=0)
    ind_tr_s7 = np.concatenate((i1, i2, i3, i4, i5, 
                                bi1, bi2, bi3, bi4, bi5, bi6), axis=0)
    
    i3d_tr_sp1 = feat_tensor[ind_tr_s1, :]
    i3d_tr_sp2 = feat_tensor[ind_tr_s2, :]
    i3d_tr_sp3 = feat_tensor[ind_tr_s3, :]
    i3d_tr_sp4 = feat_tensor[ind_tr_s4, :]
    i3d_tr_sp5 = feat_tensor[ind_tr_s5, :]
    i3d_tr_sp6 = feat_tensor[ind_tr_s6, :]
    i3d_tr_sp7 = feat_tensor[ind_tr_s7, :]
    torch.save(i3d_tr_sp1, save_dir + tensor_file[:-4] + '_tr_sp1.pth')
    torch.save(i3d_tr_sp2, save_dir + tensor_file[:-4] + '_tr_sp2.pth')
    torch.save(i3d_tr_sp3, save_dir + tensor_file[:-4] + '_tr_sp3.pth')
    torch.save(i3d_tr_sp4, save_dir + tensor_file[:-4] + '_tr_sp4.pth')
    torch.save(i3d_tr_sp5, save_dir + tensor_file[:-4] + '_tr_sp5.pth')
    torch.save(i3d_tr_sp6, save_dir + tensor_file[:-4] + '_tr_sp6.pth')
    torch.save(i3d_tr_sp7, save_dir + tensor_file[:-4] + '_tr_sp7.pth')
    
    i3d_te_sp1 = feat_tensor[bi1, :]
    i3d_te_sp2 = feat_tensor[bi2, :]
    i3d_te_sp3 = feat_tensor[bi3, :]
    i3d_te_sp4 = feat_tensor[bi4, :]
    i3d_te_sp5 = feat_tensor[bi5, :]
    i3d_te_sp6 = feat_tensor[bi6, :]
    i3d_te_sp7 = feat_tensor[bi7, :]
    torch.save(i3d_te_sp1, save_dir + tensor_file[:-4] + '_te_sp1.pth')
    torch.save(i3d_te_sp2, save_dir + tensor_file[:-4] + '_te_sp2.pth')
    torch.save(i3d_te_sp3, save_dir + tensor_file[:-4] + '_te_sp3.pth')
    torch.save(i3d_te_sp4, save_dir + tensor_file[:-4] + '_te_sp4.pth')
    torch.save(i3d_te_sp5, save_dir + tensor_file[:-4] + '_te_sp5.pth')
    torch.save(i3d_te_sp6, save_dir + tensor_file[:-4] + '_te_sp6.pth')
    torch.save(i3d_te_sp7, save_dir + tensor_file[:-4] + '_te_sp7.pth')
    
    print('*sp1--tr: ', i3d_tr_sp1.shape, '|', '--te: ', i3d_te_sp1.shape)
    print('*sp2--tr: ', i3d_tr_sp2.shape, '|', '--te: ', i3d_te_sp2.shape)
    print('*sp3--tr: ', i3d_tr_sp3.shape, '|', '--te: ', i3d_te_sp3.shape)
    print('*sp4--tr: ', i3d_tr_sp4.shape, '|', '--te: ', i3d_te_sp4.shape)
    print('*sp5--tr: ', i3d_tr_sp5.shape, '|', '--te: ', i3d_te_sp5.shape)
    print('*sp6--tr: ', i3d_tr_sp6.shape, '|', '--te: ', i3d_te_sp6.shape)
    print('*sp7--tr: ', i3d_tr_sp7.shape, '|', '--te: ', i3d_te_sp7.shape)
    
t_f1 = 'mpii_pred_48_rgb.pth'
t_f2 = 'mpii_pred_48_opt.pth'
t_f3 = 'mpii_pred_64_rgb.pth'
t_f4 = 'mpii_pred_64_opt.pth'
t_f5 = 'mpii_pred_80_rgb.pth'
t_f6 = 'mpii_pred_80_opt.pth'
t_f7 = 'mpii_pred_96_rgb.pth'
t_f8 = 'mpii_pred_96_opt.pth'
t_f9 = 'mpii_pred_48_rgb_m.pth'
t_f10 = 'mpii_pred_48_opt_m.pth'
t_f11 = 'mpii_pred_64_rgb_m.pth'
t_f12 = 'mpii_pred_64_opt_m.pth'
t_f13 = 'mpii_pred_80_rgb_m.pth'
t_f14 = 'mpii_pred_80_opt_m.pth'
t_f15 = 'mpii_pred_96_rgb_m.pth'
t_f16 = 'mpii_pred_96_opt_m.pth'

csv_f1 = 'mpii_info_list_with_subject_for_split_peter_48.csv'
csv_f2 = 'mpii_info_list_with_subject_for_split_peter_64.csv'
csv_f3 = 'mpii_info_list_with_subject_for_split_peter_80.csv'
csv_f4 = 'mpii_info_list_with_subject_for_split_peter_96.csv'

tr_te_feature_split(t_f1, csv_f1)
tr_te_feature_split(t_f2, csv_f1)
tr_te_feature_split(t_f9, csv_f1)
tr_te_feature_split(t_f10, csv_f1)

tr_te_feature_split(t_f3, csv_f2)
tr_te_feature_split(t_f4, csv_f2)
tr_te_feature_split(t_f11, csv_f2)
tr_te_feature_split(t_f12, csv_f2)

tr_te_feature_split(t_f5, csv_f3)
tr_te_feature_split(t_f6, csv_f3)
tr_te_feature_split(t_f13, csv_f3)
tr_te_feature_split(t_f14, csv_f3)

tr_te_feature_split(t_f7, csv_f4)
tr_te_feature_split(t_f8, csv_f4)
tr_te_feature_split(t_f15, csv_f4)
tr_te_feature_split(t_f16, csv_f4)
```
